```python
# büyük veri : geleneksel yöntemlerle işlenmeyen verilere deniryyy
# büyük veriyi yönetme : birden fazla bilgisayarın bir araya gelip tek bir bilgisayar gibi davranak yapılır
# veri tabanlarında tutulmayan verilere de diyoruz
```

```python
# apache hodoop : büyük veir dünyasının temelini oluşturan bir yazılım projesidir
# birden fazla pcyi bir araya getirmeye YArdımcı olur
# apache hodop bileşenleri : hodoop common ,hodoop dagıtık dosya sistemi,hodoop yarn, hodoop map reduce (ayni pcler için analizi yazılımıdır)
```

```python
# MapReduce: bilgisayar topluluğunun belirli bir amaç için birlikte hareket edebilmesini sağlayan bir programlama yazılımıdır
# input -splitting(farklı cihazlara,pc, dağıtım)-mapping(cihazlar verileri bir haritalandırdı sınıflama gibi)-shuffing(sınıflanan grubları kendi içinde ayırıd gibi)-reducing (indirgemek)-merged (toplama yuayına hazır hale getirilme)mapreducenin aşamaları
# apachehodoop faydaları : veri saklama ve işleme gücü,açık kaynak,hız,esneklik,ölçülebilirlik,hatat toleransı
```

```python
# apache spark : küme üzerinde hızlı ve genel amaçlı bilgi işleme sistemidir, hodoopun geliştirilmiş halidir,map reduce mantıgına dayanır
# java scala python ile uygulama gelişrilebilir ,spark sql
# apache spartk bileşenleri : spark core ve rdds() : hafıza yönetimi dosya saklama dosya kurtarma  
# spark sql : dataFrame lerle  yapılandırma yapılabilir
# spark MLib: makine öğrenmesi kütüphanesi 
# spark streaming: akan verinşn hataları vs işlerini ekrana aktarılır
# graphX: grafik tabanlı kullanmak için kullanılan kütüphanedir
# spark hodopun alternatifi değil mapreduce işini daha optimize yapmaya çalışan bir yazlılım
```

```python
# RDDs :dayanıklı dağınık data setler :Özetle, RDD'ler, Apache Spark platformunda dağıtılmış veri işleme işlemlerini gerçekleştirmek için kullanılan esnek ve dayanıklı veri yapılarıdır.
# hodop çalışma mnatığı : ör: evde verilen bir işi yapıp ondan sonra sana tekrardan iş yaptırmalı parça parça söyleyerk
# spark ise yapılacak işleri yapmaya başlamadan önce bekler ve senin daha iş söyleyecekmisin diye bekler ven senden  başka iş yok sözünü duyunca işlerin yapımına başlar rdds mantığı
```

```python
spark_df = spark_df.dropna()
```

```python
spark_df = spark_df.withColumn("age_kare", spark_df.age**2)
```

```python
spark_df.show(3)
```

```python
from pyspark.ml.feature import StringIndexer

stringIndexer = StringIndexer(inputCol = "churn", outputCol = "label") # 
# makine öğrenmesi için indexleme ayarı yaptı
```

```python
?stringIndexer
```

```python
# bağımlı değişkenleri belirleme ve standartlaştrıma
#
```

```python
indexed = stringIndexer.fit(spark_df).transform(spark_df)
```

```python
indexed.dtypes
```

```python
spark_df = indexed.withColumn("label", indexed["label"].cast("integer"))
```

```python
spark_df.dtypes # labeli int çevirdik
```

```python
#bagimsiz ddeğişkenleri belirleme ve standartlaştrıma
```

```python
from pyspark.ml.feature import VectorAssembler
```

```python
spark_df.columns
```

```python
bag = ["age","total_purchase", "account_manager","years","num_sites"] # bağımsız değişken için vektör oluş-şturam
```

```python
vectorAssembler = VectorAssembler(inputCols = bag, outputCol = "features")
```

```python
va_df = vectorAssembler.transform(spark_df)
```

```python
final_df = va_df.select(["features","label"]) # bağımlı ve bağımsız değişkenler
```

```python
final_df.show() # büyük veride herzaman yapılacaktır bu é!!!!
```

```python
splits = final_df.randomSplit([0.7,0.3])
train_df = splits[0]
test_df = splits[1] # test trin yapma
```

```python
train_df
```

```python
test_df
```

### ile müşteri terk modellemesi

```python
# gbm ile
```

```python
from pyspark.ml.classification import GBTClassifier
```

```python
gbm = GBTClassifier(maxIter = 10, featuresCol = "features", labelCol = "label")
```

```python
gbm_model = gbm.fit(train_df)
```

```python
y_pred = gbm_model.transform(test_df)
```

```python
y_pred
```

```python
ac = y_pred.select("label","prediction")
```

```python
ac.filter(ac.label == ac.prediction).count() / ac.count() # cv edilmemiş terk hatası
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

evaluator = BinaryClassificationEvaluator()

paramGrid = (ParamGridBuilder()
             .addGrid(gbm.maxDepth, [2, 4, 6])
             .addGrid(gbm.maxBins, [20, 30])
             .addGrid(gbm.maxIter, [10, 20])
             .build())

cv = CrossValidator(estimator= gbm, estimatorParamMaps = paramGrid, evaluator=evaluator, numFolds= 10)
```

```python
cv_model = cv.fit(train_df)
```

```python
y_pred = cv_model.transform(test_df)
ac = y_pred.select("label","prediction")
```

```python
ac.filter(ac.label == ac.prediction).count() / ac.count()
```

```python
evaluator.evaluate(y_pred)
```

```python
from pyspark.ml.classification import LogisticRegression
loj = LogisticRegression(featuresCol = "features", labelCol = 'label', maxIter=10)
loj_model = loj.fit(train_df)
y_pred = loj_model.transform(test_df)
ac = y_pred.select("label", "prediction")
ac.filter(ac.label == ac.prediction).count() / ac.count()
```

```python
from pyspark.ml.classification import DecisionTreeClassifier
dt = DecisionTreeClassifier(featuresCol = 'features', labelCol = 'label', maxDepth = 3)
dt_model = dt.fit(train_df)
y_pred = dt_model.transform(test_df)
ac = y_pred.select("label", "prediction")
ac.filter(ac.label == ac.prediction).count() / ac.count()
```
