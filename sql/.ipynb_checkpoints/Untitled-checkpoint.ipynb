```python
# JOIN işlemleri
# left join : soldaki tabloyla kesişen sağdaki değerleri birleştiriir
```

```python
%%sql
select * from Products right JOIN Categories ON Products.CategoryID=Categories.CategoryID LIMIT 3
```

```python
%%sql
select Products.ProductName,Categories.CategoryID,CategoryName  from Products right JOIN Categories ON Products.CategoryID=Categories.CategoryID LIMIT 3
```

```python
%%sql
select pr.ProductName,ca.CategoryID,ca.CategoryName  from Products as pr right JOIN Categories as ca ON pr.CategoryID=ca.CategoryID LIMIT 3
```

```python
# kısaltma yaparak daha az kod yazdık
```

```python
%%sql
select pr.ProductName,ca.CategoryID,ca.CategoryName  from Products  pr right JOIN Categories  ca ON pr.CategoryID=ca.CategoryID LIMIT 3
```

```python
# inner join iki tabloda kesişen değerleri tek tabloda yapma
```

```python
%%sql
select Orders.OrderID,Customers.CustomerName from Orders inner join Customers on Orders.CustomerID=Customers.CustomerID
```

```python
%%sql
select count(*) from (select Orders.OrderID,Customers.CustomerName from Orders inner join Customers on Orders.CustomerID=Customers.CustomerID);
```
